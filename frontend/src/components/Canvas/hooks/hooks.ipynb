Here is a guide to the most common React Hooks, categorized by their exact use cases.
## State Management

* useState: Use to track simple values like toggles, text inputs, strings, or numbers.
* useReducer: Use for complex state with multiple sub-values, or when the next state depends on the previous one.

## Side Effects

* useEffect: Use for API calls, timers, event listeners, or syncing with external systems.
* useLayoutEffect: Use to measure DOM elements (like layout width) before the browser paints the screen.

## Performance Optimization

* useMemo: Use to cache the result of an expensive mathematical calculation or data filtering.
* useCallback: Use to cache a function definition to prevent unnecessary re-renders of optimized child components.

## DOM and References

* useRef: Use to access HTML elements directly (like focusing an input) or to store a value that doesn't trigger a re-render when changed.

## Context

* useContext: Use to access global data like themes, user authentication data, or language settings without prop-drilling.

## 1. useEffect की Dependency Array का नियम
हाँ, dependency array [] में आप जो भी variables रखोगे, जब-जब उनकी value चेंज होगी, useEffect का function दोबारा run (execute) हो जाएगा।

* खाली Array []: इसके अंदर कुछ नहीं है, इसलिए यह सिर्फ पहली बार (component mount होने पर) रन होगा।
* Variables के साथ [count, userId]: जब भी count या userId में से कोई भी एक बदलेगा, यह function तुरंत फिर से रन हो जाएगा।
* बिना Array के (No array): अगर आप array देना ही भूल जाते हैं, तो component के हर सिंगल री-रेंडर पर यह रन होगा (जो कि आमतौर पर गलत माना जाता है)।

## 2. useCallback और Memoization (Caching)
हाँ, useCallback पूरी तरह से Memoization (caching) के सिद्धांत पर ही काम करता है।
लेकिन यहाँ एक छोटा सा अंतर समझना ज़रूरी है कि यह क्या कैश (cache) करता है:

* useMemo किसी function के Result (value) को कैश करता है।
* useCallback पूरे के पूरे Function Definition (स्ट्रक्चर) को ही मेमोरी में कैश कर लेता है।

यह काम कहाँ आता है?
जब React component री-रेंडर होता है, तो उसके अंदर के सारे functions दोबारा नए सिरे से बनते हैं (भले ही उनका कोड सेम हो)। अगर आप उस function को किसी Child Component में props की तरह भेज रहे हैं, तो Child Component को लगता है कि "नया function आया है" और वह भी फालतू में दोबारा रेंडर हो जाता है। useCallback इसी फालतू रेंडरिंग को रोकता है।
